```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('polymer_data.csv')
```

```python
def mic_to_numeric(x):
    if pd.isna(x):
        return np.nan
    s = str(x).strip().replace(" ", "")

    # case: >512  (or >=512 if you ever have that)
    if s.startswith(">"):
        return 9999.0

    # case: range like 256-512
    if "-" in s:
        a, b = s.split("-", 1)
        return (float(a) + float(b)) / 2.0

    # case: plain number
    return float(s)
```

```python
df["MIC_ecoli"] = df["MIC_ecoli"].apply(mic_to_numeric) / df["Mw"] # average of range, >512 --> 9999, convert to µM units
df["poly_ID"] = (df.index.to_numpy()+1).astype(int)
```

```python
df.to_csv('../polymer_db.csv')
```
