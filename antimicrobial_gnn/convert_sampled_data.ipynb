```python
# Create whole dataset
# Get list of IDs & Sequences
ids = list(df["ID"])
sequences = list(df["sequence"])
num_sequences = len(sequences)

df["sequence"].apply(lambda x: len(re.findall('[A-Z][^A-Z]*', x))).value_counts()
```

```python
for i in range(num_sequences):

    filename = "shoshana_polymers/dataset_uniform/classification/" + str(ids[i]) + "_graph.txt"

    data = []  
    data.append("MONOMERS")

    # Split sequence by capital letters (denotes different AA)
    seq = sequences[i]
    split_seq = re.findall('[A-Z][^A-Z]*', seq)
    DP = len(split_seq)

    for mon in range(DP):
        line = str(mon + 1) + " " + str(split_seq[mon])
        data.append(line)

    data.append("\n")
    data.append("BONDS")
    
    for j in range(1,DP):
        data.append(str(j) + " " + str(j + 1) + " CC")

    with open(filename, "w") as txt_file:
        for line in data:
            txt_file.write("".join(line) + "\n")  
```
